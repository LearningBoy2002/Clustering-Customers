```python
cleaned_df = df.copy()
```

```python
cleaned_df["Invoice"] = cleaned_df["Invoice"].astype(str)
cleaned_df = cleaned_df[cleaned_df["Invoice"].str.match("^\\d{6}$") == True]
cleaned_df  # only taking the proper invoice datas and leaving the others as tey are only 4 to 5
```

```python
cleaned_df["StockCode"] = cleaned_df["StockCode"].astype(str)

mask = (
    (cleaned_df["StockCode"].str.match("^\\d{5}$") == True)|
    (cleaned_df["StockCode"].str.match("^\\d{5}[a-zA-Z]+$") == True) |
    (cleaned_df["StockCode"].str.match("^PADS$") == True)
)

cleaned_df = cleaned_df[mask]

cleaned_df

#cleaning the stockcodes by matching the strings using regex(taking stock code with 5 digits, 5 digit with post alphabet and stock codes with PADS )
```

```python
cleaned_df.dropna(subset=["Customer ID"], inplace=True)# dropping the null values from "Customer ID"
```

```python
cleaned_df["SalesLineTotal"] = cleaned_df["Quantity"] * cleaned_df["Price"]
cleaned_df.head()
```

```python
aggregated_df = cleaned_df.groupby(by ="Customer ID", as_index=False)\
.agg(
    MonetaryValue =("SalesLineTotal", "sum"),
    Frequency = ("Invoice", "nunique"),
    LastInvoiceDate=("InvoiceDate", "max"),
)
aggregated_df.head()
```

```python
max_invoice_date = aggregated_df["LastInvoiceDate"].max()
max_invoice_date
```

```python
aggregated_df["Recency"] = (max_invoice_date - aggregated_df["LastInvoiceDate"]).dt.days
aggregated_df.head()
```

# **The box plot looks skewed for monetary and frequency values and there is a lot of outliers but the outliers in monetary and frequency  values are very important as they are the most spending coustomers**

```python
M_Q1 = aggregated_df["MonetaryValue"].quantile(0.25)
print(M_Q1)
M_Q3 = aggregated_df["MonetaryValue"].quantile(0.75)
print(M_Q3)
```

```python
M_IQR = M_Q3-M_Q1
print(M_IQR)
```

# **taking out the outliers as they will disturd the clustering but will do a seperate cluster for them and then merge the results**

```python
monetary_outliers_df = aggregated_df[(aggregated_df["MonetaryValue"]>(M_Q3 + 1.5* M_IQR)) | (aggregated_df["MonetaryValue"] < (M_Q1 - 1.5*M_IQR))].copy()

monetary_outliers_df.describe()
# taking only the data from 25% quantile to 75% quantile
```

```python
F_Q1 = aggregated_df["Frequency"].quantile(0.25)
print(F_Q1)
F_Q3 = aggregated_df["Frequency"].quantile(0.75)
print(F_Q3)
```

```python
F_IQR = F_Q3-F_Q1
print(F_IQR)
```

```python
frequency_outliers_df = aggregated_df[(aggregated_df["Frequency"]>(F_Q3 + 1.5* F_IQR)) | (aggregated_df["Frequency"] < (F_Q1 - 1.5*F_IQR))].copy()

frequency_outliers_df.describe()
```

```python
non_outlier_df = aggregated_df[(~aggregated_df.index.isin(monetary_outliers_df.index)) & (~aggregated_df.index.isin(frequency_outliers_df.index))]
non_outlier_df.describe()
```

# **K-Means**

```python
from sklearn.cluster import KMeans
```

```python
max_k = 12

inertia = []
k_values = range(2, max_k + 1)

for k in k_values:
  kmeans = sklearn.cluster.KMeans(n_clusters=k, random_state=42, max_iter=1000)
  kmeans.fit_predict(scaled_data_df)
  inertia.append(kmeans.inertia_)


```

```python
plt.figure(figsize=(14, 6))
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid(True)

plt.show()
```

```python
from sklearn.metrics import silhouette_score
max_k = 12

inertia = []
k_values = range(2, max_k + 1)
silhotte_scores = []


for k in k_values:
  kmeans = sklearn.cluster.KMeans(n_clusters=k, random_state=42, max_iter=1000)
  cluster_labels = kmeans.fit_predict(scaled_data_df)

  sil_score = silhouette_score(scaled_data_df, cluster_labels)
  silhotte_scores.append(sil_score)
  inertia.append(kmeans.inertia_)


```

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid(True)


plt.subplot(1, 2, 2)
plt.plot(k_values, silhotte_scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(k_values)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
overlap_indices = monetary_outliers_df.index.intersection(frequency_outliers_df.index)
overlap_indices
```

```python
monetary_only_outliers = monetary_outliers_df.drop(index=overlap_indices)
frequency_only_outliers = frequency_outliers_df.drop(index=overlap_indices)
monetary_and_frequency_outliers = monetary_outliers_df.loc[overlap_indices]

monetary_only_outliers["Cluster"] = -1
frequency_only_outliers["Cluster"] = -2
monetary_and_frequency_outliers["Cluster"] = -3

outlier_clusters_df = pd.concat([monetary_only_outliers, frequency_only_outliers, monetary_and_frequency_outliers])
outlier_clusters_df
```

```python
full_clustering_df = pd.concat([non_outlier_df, outlier_clusters_df])

full_clustering_df
```

```python
full_clustering_df["ClusterLabel"] = full_clustering_df["Cluster"].map(cluster_labels)

full_clustering_df
```
